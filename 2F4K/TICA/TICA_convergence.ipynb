### This script is to plot the TICA timescales calculated with different lag times in order to choose a resonable lag time for TICA coordinates extraction.

```python
import numpy as np
import matplotlib.pyplot as plt

%pylab inline
%config InlineBackend.figure_format='svg'
```

```python
# Load the timescales
data = np.load('./2F4K_TICA_timescales_all.npz')
print(list(data.keys()))
```

```python
# Check the dimensions of data
timescales = data['timescales']
lags = data['lags']

print(np.shape(timescales))
print(np.shape(lags))
```

```python
# Convert the unit from step to real time
lags = lags/5000.0
timescales = timescales/5000.0
```

```python
plt.plot(lags[0:22],timescales[0:22,0:10],linewidth = 0.7)
plt.axis([0,0.1,0.001,10])

x = np.arange(0.001,0.2,0.001)
plt.plot(x,x,color='black')
semilogy() 

plt.xlabel('lag time $\\tau$ ($\\mu$s)')
plt.ylabel('TICA timescale $t_{i}$ ($\\mu$s)')

axvline(x=0.02, color='blue', linestyle='-')
```

### In this case, we choose 0.02 microsecond (i.e. 100 steps) as the lag time for further calculation
