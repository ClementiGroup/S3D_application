### This script is to plot the MSM implied timescales calculated with different lag times in order to choose a resonable lag time for HMM construction.

```python
import matplotlib.pyplot as plt
from pyemma import msm
import numpy as np

%pylab inline
%config InlineBackend.figure_format='svg'
```

```python
# Load the discrete trajectory
data = np.load('./2F4K_MSM_10TICA_clusters_1000.npz')
print(list(data.keys()))

n_clusters = data['n_clusters']
discrete_traj = data['micro_membership']
centers = data['centers']
```

```python
# Create the list af lag times
MSMlags = np.array([1])
for i in range(1,4,1):
    nmin = 10**(i)
    nmax = 10**(i+1)
    dn = 10**(i)
    MSMlags1 = np.arange(nmin,nmax,dn)
    MSMlags = np.concatenate([MSMlags,MSMlags1])

nlagsMSM = np.shape(MSMlags)[0]
print('number of different lag times = ' + str(nlagsMSM))
print('list of lag times for MSM     = ' + str(MSMlags))
```

```python
# Load the transition matrices for different lag times
a = np.load('2F4K_MSMtransition_matrix_10TICA_1000.npz')

# Calculate the implied timescales
msm_t = []
n_timescales = 20

for i in range(nlagsMSM):
    mm = msm.markov_model(a[str(i)])
    msm_t.append(mm.timescales()[0:n_timescales]*MSMlags[i])
    
print('done loading set of MSM!')
```

```python
# Scale the timescales to real time unit
MSMlags = np.array(MSMlags,dtype = float)
msm_t = np.array(msm_t,dtype = float)

MSMlags = MSMlags/5000.0
msm_t = msm_t/5000.0
```

```python
# Plot and choose lag time for HMM construction
plt.plot(MSMlags,msm_t[:,0:10],'-',linewidth = 0.7)
plt.axis([0,0.05,0.001,10])

x = np.arange(0.001, 2.0,0.001)
plt.plot(x,x,color='black')
semilogy() 

plt.xlabel('lag time $\\tau$ ($\\mu$s)')
plt.ylabel('ITS $t_{i}$ ($\\mu$s)')

axvline(x=0.01, color='blue', linestyle='-')
```

### In this case, we choose 0.01 microsecond (i.e. 50 steps) as the lag time for further calculation
